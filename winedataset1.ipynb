# Bivariate analysis

```python
plt.scatter(df['Class'],df['Alcohol'])
sns.set(style='whitegrid')
plt.title('Scatter plot')

plt.show()
```

```python
sns.violinplot(df['Class'],df['Alcohol'])
sns.set(style='whitegrid')
plt.title('Scatter plot')
plt.xlabel('Class')
plt.ylabel('Alcohol')

plt.show()
```

```python
sns.swarmplot(df['Class'],df['Alcohol'])
sns.set(style='whitegrid')
plt.title('Scatter plot')
plt.xlabel('Class')
plt.ylabel('Alcohol')

plt.show()
```

```python
sns.boxplot(df['Class'],df['Alcohol'])
sns.set(style='whitegrid')
plt.title('Scatter plot')
plt.xlabel('Class')
plt.ylabel('Alcohol')

plt.show()
```

```python
corr=df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```
